### REQUIRED LIBRARY IMPORTED

```python
import pandas as pd
df=pd.read_csv("Toronto_collision_raw.csv")
```

### DATA CLEANING

```python
df=df.drop(columns = {"OBJECTID", "EVENT_UNIQUE_ID", "HOOD_158", "LONG_WGS84", "LAT_WGS84", "x", "y", "DIVISION", "OCC_DATE", "PD_COLLISIONS"})
```

```python
df.columns = ["accident_month", "accident_day", "accident_year", "accident_hour", "fatalities", "injuries", "collisions", "area", "car", "motorcycle", "passengers", "bicycle", "pedestrians"]
```

```python
df
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df[df['pedestrians']== "N/R"]
```

```python
df["motorcycle"] = df["motorcycle"].replace("N/R", "not_recorded")
df["passengers"] = df["passengers"].replace("N/R", "not_recorded")
df["bicycle"] = df["bicycle"].replace("N/R", "not_recorded")
df["pedestrians"] = df["pedestrians"].replace("N/R", "not_recorded")
df["car"] = df["car"].replace("N/R", "not_recorded")
df
```

```python
df["area"].unique()
df["area"] = df["area"].replace("NSA", "not_recorded")
```
